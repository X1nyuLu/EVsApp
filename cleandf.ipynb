```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('attribution.csv')
```

```python
df.head()
```

```python
for i in range(len(df)):
    tmp = [str(df.iloc[i, j]) for j in range(1, 11) if not pd.isna(df.iloc[i, j])]
    tmp_str = '\n'.join(tmp)
    df.loc[i, 'assignment'] =  tmp_str
    df.loc[i, 'peak'] = int(df.iloc[i, 0].replace('$cm^{-1}$', ''))
```

```python
df.head()
```

```python
new_df = df[['peak', 'assignment']].drop_duplicates()
```

```python
new_df.head()
```

```python
new_df.to_csv('cleaned_attribution.csv', index=False)
```
